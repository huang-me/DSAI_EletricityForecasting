# Plot datas

```python
# import modules
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# all file paths
path_14_17 = os.path.join('datas', 'Taipower_20140101_20171231.csv')
path_18_190131 = os.path.join('datas', 'TaiPower_20180101_20190131.csv')
path_19_201031 = os.path.join('datas', 'TaiPower_20190101_20201031.csv')
```

```python
# read files
data_14_17 = pd.read_csv(path_14_17)
data_18_190131 = pd.read_csv(path_18_190131)
data_19_201031 = pd.read_csv(path_19_201031)
```

##### Preprocessing of datas 2014~2017

```python
data_14_17.head()
```

```python
data_14_17['peak'] = data_14_17['備轉容量(MW)'] / data_14_17['備轉容量率(%)'] * 100
data_14_17['peak'] = data_14_17['peak'].round()
data_14_17.head()
```

```python
data_14_17['date'] = pd.to_datetime(data_14_17['日期'], format="%Y/%m/%d").dt.strftime("%Y%m%d")
useless_col = ['備轉容量率(%)', '備轉容量(MW)', '日期']
data_14_17 = data_14_17.drop(useless_col, axis=1)
data_14_17.head()
```

##### Extract datas 2018

```python
data_18_190131.head()
```

```python
data_18 = data_18_190131[data_18_190131['日期'] <= 20181231]
data_18.head()
```

##### Extract data 2019

```python
data_19_201031.head()
```

```python
data_19 = data_19_201031[data_19_201031['日期'] <= 20191231]
data_19.head()
```

##### Delete unneeded column

```python
col = ['日期', '尖峰負載(MW)']
data_18 = data_18[col]
data_19 = data_19[col]
```

##### Rename peak & time

```python
rename_dict = {'日期':'date', '尖峰負載(MW)':'peak'}
data_18.rename(columns=rename_dict, inplace=True)
data_19.rename(columns=rename_dict, inplace=True)
data_18.head(), data_19.head()
```

##### Concatenate all data

```python
all_data = pd.concat((data_14_17, data_18, data_19))
all_data.head(), all_data.tail(), len(all_data)
```

```python
plt.figure(figsize=(20,4))
x_axis = pd.to_datetime(all_data['date'], format='%Y%m%d')
plt.plot(x_axis, all_data['peak'])
plt.title('Peak')
```

```python
plt.figure(figsize=(20,4))
x_axis = range(0, 365)
plt.plot(x_axis[:len(all_data[:365])], all_data[:365]['peak'], '-')
plt.plot(x_axis[:len(all_data[365*1:365*2])], all_data[365*1:365*2]['peak'], '-')
plt.plot(x_axis[:len(all_data[365*2:365*3])], all_data[365*2:365*3]['peak'], '-')
plt.plot(x_axis[:len(all_data[365*3:365*4])], all_data[365*3:365*4]['peak'], '-')
plt.plot(x_axis[:len(all_data[365*4:365*5])], all_data[365*4:365*5]['peak'], '-')
plt.plot(x_axis[:len(all_data[365*5:])], all_data[365*5:]['peak'], '-')
plt.title('Peak Loads')
plt.legend(['2014', '2015', '2016', '2017', '2018', '2019'])
plt.show()
```
